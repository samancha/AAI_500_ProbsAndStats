The `Houses` data file ([http://stat4ds.rwth-aachen.de/data/Houses.dat](http://stat4ds.rwth-aachen.de/data/Houses.dat)) at the bookâ€™s website lists the selling price (thousands of dollars), size (square feet), tax bill (dollars), number of bathrooms, number of bedrooms, and whether the house is new (1 = yes, 0 = no) for 100 home sales in Gainesville, Florida. Letâ€™s analyze the selling prices. 

(a) Construct a frequency distribution and a histogram. Describe the shape.  
(b) Find the percentage of observations that fall within one standard deviation of the mean. Why is this not close to 68%?  
(c) Construct a box plot, and interpret.   
(d) Use descriptive statistics to compare selling prices according to whether the house is new. 




**Answer:**

`(a)`

The shape is a high skwed right plot with a few highly priced houses as outliers and the highly priced houses also are less in frequency.  

```python
data_url = 'https://stat4ds.rwth-aachen.de/data/Houses.dat'
house_data = pd.read_csv(data_url, delim_whitespace=True)

# Reading data and displaying raw data fram
df = pd.DataFrame(house_data)
display(df)

# display freq (is a type Series)
freq = df['price'].value_counts()
display(freq.to_frame())

# # Creating the plot 
sns.histplot(data=freq.index.values, bins='auto')

# Adding labels, titles, and showing my plot
plt.xlabel("House price (Thousands of dollars)")
plt.ylabel("Frequency")
plt.title("Fequence Histogram of Housing Price Dataset")
plt.show()
```

`(b)`

The value of 68% corresponds to the percentage of observations within one standard deviation for a normally distributed dataset. Since the housing price data has a few major outliers, the percentage within one standard deviation deviates from 68%.

```python
mean_income = df['price'].mean()
std_income = df['price'].std()
lower_bound = mean_income - std_income
upper_bound = mean_income + std_income

# Recreates array to only include values within the range and I find the percentage 
within_one_std = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
percentage_within_one_std = (len(within_one_std) / len(df)) * 100

print(f"Mean : {mean_income:.2f}")
print(f"Range : {lower_bound:.2f}-{upper_bound:.2f}")
print(f"Percentage within one standard deviation: {percentage_within_one_std:.2f}%")
print(f"Standard Deviation: {std_income:.2f}")
```

`(c)`
 Construct a box plot, and interpret.

I can see that there are a lot of expensive houses because of the amount of higher end outliers. 

```python
# Creating a box plot with housing prices
sns.boxplot(data=df, x='price', color='red')
plt.xlabel('Housing Price (thousands of dollars)')
plt.ylabel('Houses')
plt.title('Box Plot of Housing Prices')
plt.show()
```

`(d)` The mean is alot more expensive for selling new houses and there are a lot more non-new houses for sale.
